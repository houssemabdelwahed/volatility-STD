```python
import pandas as pd
import yfinance as yf
import numpy as np
from datetime import datetime
```

```python
stockstartdate='2020-01-01'
today=datetime.today().strftime('%Y-%m-%d')
today
```

```python
stock = yf.download("BTC-USD", start=stockstartdate, end=today, interval='1d')
```

```python
stock
```

```python
stock['Log returns'] = np.log(stock['Close']/stock['Close'].shift())
```

```python
stock
```

```python
stock['Log returns'].std()
```

```python
stock['STD'] = stock['Log returns'].std()
stock
```

```python
volatility = stock['Log returns'].std()*365**.5
volatility
```

```python
import matplotlib.pyplot as plt
```

```python
str_vol = str(round(volatility, 4)*100)
fig, ax = plt.subplots()
stock['Log returns'].hist(ax=ax, bins=50, alpha=0.6, color='b')
ax.set_xlabel('Log return')
ax.set_ylabel('Freq of log return')
ax.set_title('BTC-USD volatility: ' + str_vol + '%')
```
